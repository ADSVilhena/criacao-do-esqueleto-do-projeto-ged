```python
import cv2
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread('witcher.jpeg')
img2 = cv2.imread('ffxv.jpeg')
img3 = cv2.imread('dgi.jpeg')
```

```python
img1Gray = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
img2RGB = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
img3HSV = cv2.cvtColor(img3, cv2.COLOR_BGR2HSV)
```

```python
cv2.imwrite('img1gray.jpeg', img1Gray)
cv2.imwrite('img2RGB.jpeg', img2RGB)
cv2.imwrite('img3HSV.jpeg', img3HSV)
```

```python
img1 = cv2.imread('img1gray.jpeg')
img2 = cv2.imread('img2RGB.jpeg')
img3 = cv2.imread('img3HSV.jpeg')
```
